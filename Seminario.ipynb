```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv("victimas-accionar-represivo-ilegal.csv", encoding="cp1252")
```

```python
def get_age(value):
    try:
        value = int(value[:2])
    except ValueError:
        value = 0
    return value
```

```python
dt.head(4)
```

```python
edades = dt['edad_al_momento_del_hecho'].apply(get_age)
dt['edad_al_momento_del_hecho'] = dt['edad_al_momento_del_hecho'].apply(get_age)
```

```python
print("Cantidad de secuestrados", len(dt))
print("Cantidad embarazadas: ", dt['embarazo'].count())
print("Edad promedio: ", round(edades.mean(), 2))
print("Edad más sufrida: ", round(edades.mode()[0], 2))
#print("Edades: ", edades.value_counts())
```

```python
edades.hist(bins=10, grid=True, figsize=(12,8), color='c', rwidth=0.9)
```
